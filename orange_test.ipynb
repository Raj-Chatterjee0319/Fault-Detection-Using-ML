```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train XGBoost
xgb = XGBClassifier(n_estimators=100, learning_rate=0.1, random_state=42)
xgb.fit(X_train, y_train)

# Train SVM with probability enabled
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)

# Get model probability predictions
xgb_probs = xgb.predict_proba(X_test)  # XGBoost probabilities
svm_probs = svm.predict_proba(X_test)  # SVM probabilities

# Hybrid Model: Averaging probabilities
final_probs = (xgb_probs + svm_probs) / 2
final_preds = np.argmax(final_probs, axis=1)  # Convert probabilities to class labels

# Evaluate Hybrid Model
hybrid_acc = accuracy_score(y_test, final_preds)
print(f"Hybrid Model (SVM + XGBoost) Accuracy: {hybrid_acc:.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import joblib
from xgboost import XGBClassifier
from sklearn.svm import SVC

def predict_fault(time, V_dc, I_dc, P_dc, I_Grid, P_Grid, T_PV, G):
    """
    Function to predict faults using a hybrid SVM + XGBoost model.

    Args:
        time (float): Time value
        V_dc (float): DC Voltage
        I_dc (float): DC Current
        P_dc (float): DC Power
        I_Grid (float): Grid Current
        P_Grid (float): Grid Power
        T_PV (float): PV Temperature
        G (float): Irradiance

    Returns:
        tuple: (is_fault, predicted_label)
              - is_fault (bool): True if fault is detected
              - predicted_label (str): Specific fault label if fault is detected
    """

    # Create input DataFrame with all required features
    input_data = pd.DataFrame({
        'time': [time],
        'V_dc': [V_dc],
        'I_dc': [I_dc],
        'P_dc': [P_dc],
        'I_Grid': [I_Grid],
        'P_Grid': [P_Grid],
        'T_PV': [T_PV],
        'G': [G]
    })

    # Load the saved models and scaler
    try:
        scaler = joblib.load('scaler.joblib')
        xgb_model = joblib.load('xgb_model.joblib')
        svm_model = joblib.load('svm_model.joblib')
    except FileNotFoundError:
        print("Error: Model files not found. Please train and save the models first.")
        return False, None

    # Preprocess data
    scaled_features = scaler.transform(input_data[['time', 'V_dc', 'I_dc', 'P_dc',
                                                  'I_Grid', 'P_Grid', 'T_PV', 'G']])

    # Create DataFrame with scaled values
    scaled_df = pd.DataFrame(scaled_features, columns=['time', 'V_dc', 'I_dc', 'P_dc',
                                                      'I_Grid', 'P_Grid', 'T_PV', 'G'])

    # Make predictions using both models
    xgb_probs = xgb_model.predict_proba(scaled_df)
    svm_probs = svm_model.predict_proba(scaled_df)

    # Hybrid approach: Average probabilities
    final_probs = (xgb_probs + svm_probs) / 2
    predicted_class = np.argmax(final_probs, axis=1)[0]

    # Determine if it's a fault
    is_fault = predicted_class != 0  # Assuming 0 is the "No Fault" class

    # Get fault label based on predicted class
    fault_labels = {0: 'No Fault', 1: 'Fault 1', 2: 'Fault 2',
                   3: 'Fault 3', 4: 'Fault 4'}  # Update labels as per your dataset
    predicted_label = fault_labels.get(predicted_class, 'Unknown')

    return is_fault, predicted_label

# Example usage:
if __name__ == "__main__":
    # Sample input data
    time = 1000
    V_dc = 486
    I_dc = 172
    P_dc = 83953
    I_Grid = 2.5
    P_Grid = 81
    T_PV = 77
    G = 580

    is_fault, predicted_label = predict_fault(time, V_dc, I_dc, P_dc,
                                             I_Grid, P_Grid, T_PV, G)

    if is_fault:
        print(f"Fault detected: {predicted_label}")
    else:
        print("No fault detected")
```
